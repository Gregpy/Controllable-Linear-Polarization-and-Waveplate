```python
import matplotlib.pyplot as plt
import scipy.constants as pc
import numpy as np

%matplotlib inline
```

```python
lm = 800*10**-9 #wavelength of laser, m
w = 2*np.pi*pc.c/lm  #angular frequency, omega, of laser, Hz
lp = 0*2*np.pi #angle of laser linear polarization, measured in radians from horizontal axis
h = 0.125*2*np.pi #angle of fast axis of the waveplate, measured in radians from horizontal axis
n= np.pi/2 #phase retardation of the waveplate (e.g. for a QWP, n=np.pi/2, for a HWP, n=np.pi)
t = np.linspace(0,2*np.pi/w) #time wave propagates for a 2*pi cycle, s
```

```python
plt.figure(figsize = (10,6))

plt.subplot(121, aspect = 'equal')
xlp = np.real(np.exp(-1.j*w*t)*np.cos(lp))
ylp = np.real(np.exp(-1.j*w*t)*np.sin(lp))
plt.plot(xlp,ylp)
plt.xlabel('$E_{90^\circ}$')
plt.ylabel('$E_{0^\circ}$')
plt.title('Polarization before Waveplate')
plt.axis([-1,1,-1,1])

plt.subplot(122, aspect = 'equal')
xeeo = np.real(np.exp(-1.j*w*t)*(np.cos(lp)*(np.exp(1.j*n/2)*(np.cos(h))**2+np.exp(-1.j*n/2)*(np.sin(h))**2)+(-np.exp(-1.j*n/2)+np.exp(1.j*n/2))*np.cos(h)*np.sin(h)*np.sin(lp)))
yeeo = np.real(np.exp(-1.j*w*t)*((-np.exp(-1.j*n/2)+np.exp(1.j*n/2))*np.cos(h)*np.cos(lp)*np.sin(h)+(np.exp(-1.j*n/2)*(np.cos(h))**2+np.exp(1.j*n/2)*(np.sin(h))**2)*np.sin(lp)))
plt.plot(xeeo,yeeo)
plt.xlabel('$E_{90^\circ}$')
plt.ylabel('$E_{0^\circ}$')
plt.title('Polarization after Waveplate')
plt.axis([-1,1,-1,1])

plt.tight_layout()
plt.show()
```
